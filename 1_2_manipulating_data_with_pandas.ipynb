## 1. Create a 4 column DataFrame with 10 rows, the first column being a date field and the rest numbers.

```
df = pd.DataFrame({"col1": pd.Series([], dtype="datetime64[M]"),
                  "col2": pd.Series([], dtype="float"),
                  "col3": pd.Series([], dtype="float"),
                  "col4": pd.Series([], dtype="float")})
```

## 2. Fill the first column with the first day of each month for 3 years (for example: 1/1/2018, 2/1/2018)

```
dates = pd.date_range(start="2017-12-31", periods=36, freq="M")
```

```
df["col1"] = pd.date_range(start="2017-12-31", periods=36, freq="M").shift(periods=1, freq="D")
```

## 3. Fill the next 2 columns with random numbers.

```
df["col2"] = np.random.random(36)
```

```
df["col3"] = np.random.random(36)
```

## 4. Fill the 4th column with the difference of the first 2 data columns (for example: Col3 - Col2)

```
df["col4"] = df["col3"] - df["col2"]
```
